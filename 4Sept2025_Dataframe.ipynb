# Dataframe

### Dataframe is used to arrange Data in Structured way

### Convert Dictionary into Dataframe

```python
dict1={
    "EmpID":[101,102,103,104,105],
    "EmpName":['Ankita','Pragati','Jyoti','Sujata','Dipika'],
    "EmpSalary":[20000,35000,30000,15000,25000]
}
```

```python
dict1
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(dict1)
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

### Convert List to a Dataframe

```python
l1=[1,2,3,4,5,6]
l2=['A','B','C','D','E','F']
l3=[12,15,17,16,18,19]
l4=[45,67,78,89,93,53]
```

```python
df1=pd.DataFrame([l1,l2,l3,l4]).T
```

```python
df1
```

```python
df1.shape
```

```python
df1.columns
```

```python
df1.columns=["sid",'sname','sage','smarks']
```

```python
df1
```

### Data Subsetting

#### Used to remove small part from Dataset

## iloc
#### Syntax= df.iloc[start_row_index:end_row_index,start_column_index:end_column_index]

```python
df.iloc[10:21,3:5]
```

## loc
#### Syntax= df.loc[start_row_index:end_row_index,['column1','column2']]

```python
df.loc[10:20,['STATE','PROFIT']]
```

### Data Filtering

#### Find out companies who have profit more than 1.5 L & are from Florida state

```python
df[(df['PROFIT']>150000) & (df['STATE']=='Florida')]
```

```python
### Find out Data for those companies who have PROFIT between 1 to 1.5L and which are from california or florida
```

```python
df[((df['PROFIT']>100000) & (df['PROFIT']<150000)) & ((df['STATE']=='California') | (df['STATE']=='Florida'))]
```
